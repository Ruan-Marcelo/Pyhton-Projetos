```python
import math

from turtle import *

def hearta(R):
    return 12*math.sin(R)**3
def heartb(R):
    return 12*math.cos(R)-5*\
    math.cos(2*R)-2*\
                math.cos(3*R)-\
                math.cos(4*R)
speed(0)
bgcolor("black")
for i in range(10000):
    goto(hearta(i)*20, heartb(i)*2)
    for j in range(5):
        color("#f73487")
        goto(0,0)
done()
```
